```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
pathrooth = '/home/alb/Scrivania/Quantum/Projects/BornMachine/data/'

pathleaf_MMD07_Loss = 'MMDTraining4qb400italpha007_MMD_history.csv'
pathleaf_MMD10_Loss = 'MMDTraining4qb400italpha010_MMD_history.csv'
pathleaf_KLGauss07_Loss = 'KLGaussTraining4qb400italpha007_KL_history.csv'
pathleaf_KLGauss10_Loss = 'KLGaussTraining4qb400italpha010_KL_history.csv'
pathleaf_KLHypercube07_Loss = 'KLHypercubeTraining4qb400italpha007_KL_history.csv'
pathleaf_KLHypercube10_Loss = 'KLHypercubeTraining4qb400italpha010_KL_history.csv'

pathleaf_MMD07_TV = 'MMDTraining4qb400italpha007_TV_history.csv'
pathleaf_MMD10_TV = 'MMDTraining4qb400italpha010_TV_history.csv'
pathleaf_KLGauss07_TV = 'KLGaussTraining4qb400italpha007_TV_history.csv'
pathleaf_KLGauss10_TV = 'KLGaussTraining4qb400italpha010_TV_history.csv'
pathleaf_KLHypercube07_TV = 'KLHypercubeTraining4qb400italpha007_TV_history.csv'
pathleaf_KLHypercube10_TV = 'KLHypercubeTraining4qb400italpha010_TV_history.csv'
```

```python
MMD07_Loss = pd.read_csv(pathrooth + pathleaf_MMD07_Loss, header= None)
MMD07_TV = pd.read_csv(pathrooth + pathleaf_MMD07_TV, header= None)
MMD10_Loss = pd.read_csv(pathrooth + pathleaf_MMD10_Loss, header= None)
MMD10_TV = pd.read_csv(pathrooth + pathleaf_MMD10_TV, header= None)

KLGauss07_Loss = pd.read_csv(pathrooth + pathleaf_KLGauss07_Loss, header= None)
KLGauss07_TV = pd.read_csv(pathrooth + pathleaf_KLGauss07_TV, header= None)
KLGauss10_Loss = pd.read_csv(pathrooth + pathleaf_KLGauss10_Loss, header= None)
KLGauss10_TV = pd.read_csv(pathrooth + pathleaf_KLGauss10_TV, header= None)

KLHypercube07_Loss = pd.read_csv(pathrooth + pathleaf_KLHypercube07_Loss, header= None)
KLHypercube07_TV = pd.read_csv(pathrooth + pathleaf_KLHypercube07_TV, header= None)
KLHypercube10_Loss = pd.read_csv(pathrooth + pathleaf_KLHypercube10_Loss, header= None)
KLHypercube10_TV = pd.read_csv(pathrooth + pathleaf_KLHypercube10_TV, header= None)
```

```python
names = ['MMD07', 'MMD10', 'KLGauss07', 'KLGauss10', 'KLHypercube07', 'KLHypercube10']
TVdf = pd.concat([MMD07_TV, MMD10_TV, KLGauss07_TV, KLGauss10_TV, KLHypercube07_TV, KLHypercube10_TV ], axis = 1)
Lossdf = pd.concat([MMD07_Loss, MMD10_Loss, KLGauss07_Loss, KLGauss10_Loss, KLHypercube07_Loss, KLHypercube10_Loss ], axis = 1)
TVdf.columns = names
Lossdf.columns = names
```

```python
plt.style.use('seaborn-darkgrid')
palette = plt.get_cmap('Set1')
 
num=0
for column in TVdf:
    num+=1
    plt.plot(TVdf[column], marker='', color=palette(num), linewidth=1, alpha=0.9, label=column)


plt.legend(loc=2, ncol=2)
 
plt.title("TV comparison", loc='left', fontsize=12, fontweight=0, color='orange')
plt.xlabel("Epoch")
plt.ylabel("TV")

plt.show()
```

```python
plt.style.use('seaborn-darkgrid')
palette = plt.get_cmap('Set1')
 
num=0
for column in Lossdf:
    num+=1
    plt.plot(Lossdf[column], marker='', color=palette(num), linewidth=1, alpha=0.9, label=column)


plt.legend(loc=2, ncol=2)
 
plt.title("Loss comparison", loc='left', fontsize=12, fontweight=0, color='orange')
plt.xlabel("Epoch")
plt.ylabel("Loss")
#plt.ylim(top=0.25)  
plt.show()
```
